```python
matrix = np.array([[2, 1], [4, 3]])
inverse_matrix = np.linalg.inv(matrix)
print(inverse_matrix)
```

```python
result = np.matmul(matrix, inverse_matrix)
print(result)
```

```python
result = matrix @ inverse_matrix
result
```

```python
import numpy as np
import matplotlib.pyplot as plt
fre=int(input("freq ="))
x = np.linspace(0, 2 * np.pi*fre, 1000)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.title('Sine Wave')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fre=int(input("freq ="))
x = np.linspace(0, 2 * np.pi*fre, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1+y2)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.title('Sine Wave')
plt.show()
```
